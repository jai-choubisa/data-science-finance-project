# Finance Data Project 

```python
from pandas_datareader import data, wb
import pandas as pd
import numpy as np
import datetime
```

## Data

Getting data using pandas datareader. We will get stock information for the following banks:
*  Bank of America
* CitiGroup
* Goldman Sachs
* JPMorgan Chase
* Morgan Stanley
* Wells Fargo

** Getting the stock data from Jan 1st 2006 to Jan 1st 2016 for each of these banks. **

```python
start = datetime.datetime(2006,1,1)
end = datetime.datetime(2016,1,1)
# BAC = data.DataReader("BAC",'google',start,end)
#google api expired no loger working
```

# Step 1: Data Import

```python
# Bank of America
BAC = data.DataReader("BAC", 'yahoo', start, end)

# CitiGroup
C = data.DataReader("C", 'yahoo', start, end)

# Goldman Sachs
GS = data.DataReader("GS", 'yahoo', start, end)

# JPMorgan Chase
JPM = data.DataReader("JPM", 'yahoo', start, end)

# Morgan Stanley
MS = data.DataReader("MS", 'yahoo', start, end)

# Wells Fargo
WFC = data.DataReader("WFC", 'yahoo', start, end)
```

** Creating a list of the ticker symbols (as strings) in alphabetical order **

```python
tickers = ['BAC','C','GS','JPM','MS','WFC']
```

** Using pd.concat to concatenate the bank dataframes together to a single data frame called bank_stocks. To be used for column names of above dataframes **

```python
bank_stocks = pd.concat([BAC,C,GS,JPM,MS,WFC],axis = 1,keys=tickers)
```

```python
bank_stocks.columns.names = ['Bank Ticker','Stock Info']
```

** Check the head of the bank_stocks dataframe.**

```python
bank_stocks.head()
```

# Step 2: Data Analysis


** What is the max Close price for each bank's stock throughout the time period?**

```python
bank_stocks.xs(key='Close',level='Stock Info',axis=1).max()
```

** Lets Create a new empty DataFrame called returns. This dataframe will contain the returns for each bank's stock. returns are typically defined by:**

$$r_t = \frac{p_t - p_{t-1}}{p_{t-1}} = \frac{p_t}{p_{t-1}} - 1$$

```python
returns = pd.DataFrame()
```

** We can use pandas pct_change() method on the Close column to create a column representing this return value. Create a for loop that goes and for each Bank Stock Ticker creates this returns column and set's it as a column in the returns DataFrame.**

```python
for tick in tickers:
    returns[tick+' Return'] = bank_stocks[tick]['Close'].pct_change()
returns.head()
```

** Lets Create a pairplot using seaborn of the returns dataframe.**

```python
import seaborn as sns
sns.pairplot(returns[1:])
```

From above plot we can see comparison of returns between all banks with each other. It seens BAC Returns rise highest while comparing with other banks and MS Returns rise lowest while comparing with all other banks.

** Lets check data for bank stocks when they had the best and worst single day returns. **

```python
returns.idxmin()
```

On 20 January 2009, 4 bank stocks BAC,GS,JPM and WFC all have the lowest returns of all the returns dataset.

** Lets see highest return date **

```python
returns.idxmax()
```

It seens C and GS have highest returns on same day 24 November 2008.

** Lets get the standard deviation of the returns, which stock would you classify as the riskiest over the entire time period? Which would you classify as the riskiest for the year 2015?**

```python
returns.std()
```

From above returns Standard deviation, we could see Citygroup have highest standard deviation which means it is riskiest over entire period.

```python
returns.loc['2015-01-01':'2015-12-31'].std()
```

from above return values we can see Morgan Stanley Bank was riskiest (Bank of America very close) in 2015 year.

# Step 3: Data Visualization

```python
sns.set_style('whitegrid')
returns_2015 = returns.loc['2015-01-01':'2015-12-31']
sns.distplot(returns_2015['MS Return'],bins=50,color='Green',kde=False)
```

From above plot we can see Morgan Stanley bank returns for 2015 year. Most of the returns came between -0.02 to 0.02 and largest number of returns came around 0.0 .

## Moving Averages

Let's analyze the moving averages for these stocks in the year 2008. Lets check for Bank of America.

```python
plt.figure(figsize = (16,6))
BAC['Close'].loc['2008-01-01':'2009-01-01'].rolling(window=30).mean().plot(label='30 Day Avg')
BAC['Close'].loc['2008-01-01':'2009-01-01'].plot(label='BAC CLOSE')
plt.legend()
```

From above stock 'close' data we can see that Bank of America stocks 'close' was decreasing in 2008 and from around 45 in February it decreased to less than 15 at the end of year.

** Lest create a heatmap of the correlation between the stocks Close Price.**

```python
sns.heatmap(bank_stocks.xs(key='Close',level='Stock Info',axis = 1).corr(),annot = True)
```

from above plot we can see 
1. Bank of America and Citygroup banks 'Close' stocks are highly corelate. 
2. Also Morgan Stanley is also closely related to both of above banks 'Close' stock.
3. JPMorgan Chase is highly corelated with Wells Fargo Banks 'Close' stock.

** Above corelated banks can be grouped together and visualized using clustermap **

```python
sns.clustermap(bank_stocks.xs(key='Close',level='Stock Info',axis = 1).corr(),annot = True)
#We can see below corelated banks stocks are clustered together.
```
